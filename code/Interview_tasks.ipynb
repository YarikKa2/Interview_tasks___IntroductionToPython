#### Найти пересечение без дубликатов

___

Пользователь с клавиатуры (через функцию `input()`) вводит через запятую и пробел две последовательности чисел (некоторые идентификаторы товаров/услуг, просмотренных клиентом A и клиентом B на платформе «Авито»).

Необходимо конвертировать каждый ввод пользователя в отдельный список, состоящий из целых чисел, а также составить новый список, состоящий из элементов пересечения двух исходных. Дубликаты необходимо удалить, а результирующий список — вывести на экран.

**Важно!** Если пересечение пустое, то на экран должен выводиться пустой список.

Если ввод пользователя невозможно конвертировать в список из целых чисел, возникает ошибка. Необходимо отлавливать эту ошибку и выводить на экран фразу «Некорректный ввод».

Для поиска пересечений в списках напишите функцию `pure_intersection()`, которая будет возвращать список с уникальными элементами пересечения.


**Пример №1:**
```

```

**Подсказка:**

Для определения пересечения преобразуйте списки во множества и воспользуйтесь методом `intersection()`.

**Критерии оценивания:**

- Задача решена корректно, выполнены все условия (**4 балла**):
    - каждый ввод пользователя конвертируется в список из идентификаторов;
    - функция для определения пересечения корректно работает для любого ввода пользователя, обозначенного в условиях задачи;
    - решение сопровождено комментариями.
- Задача решена корректно, однако не выполнено одно из условий (**2 балл**):
    - ввод пользователя не конвертируется в список;
    - функция для определения пересечения работает только для ограниченного множества входных данных в рамках условия задачи;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): представлен некорректный алгоритм определения пересечения элементов.

````python
input_string1 = input('Введите 1-ую последовательность идентификаторов: ')
input_string2 = input('Введите 2-ую последовательность идентификаторов: ')

# создаем функцию которая на вход получает две строки str_1 и str_2
def pure_intersection(str_1, str_2):
  
  
  # создаем конструкцию try-except , если будут поданы данные, неконвертируемые в int()
  try:
    
    # разделяем 1-ую строку на подстроки; конвертируем в int
    str_1 = str_1.split(', ') 
    str_1 = list(map(lambda x: int(x), str_1)) 

    # разделяем 2-ую строку на подстроки; конвертируем в int
    str_2 = str_2.split(', ')
    str_2 = list(map(lambda x: int(x), str_2))

 # если данные некорректны, выводим надпись "Некорректный ввод"
  except ValueError: 
    
    return print('Некорректный ввод')


  # оцениваем пересечение между множеством из первого списка (str_1) и из второго списка (str_2)
  return list(set(str_1).intersection(set(str_2)))


pure_intersection(input_string1, input_string2)
````

## Задание №2

#### Найти минимум и максимум

**Компания:** «Авито»

**Уровень:** Лёгкий

___

Пользователь подаёт на вход программе действительные числа через пробел с клавиатуры (через функцию `input()`).

Он привык пользоваться Excel, поэтому отделяет дробную часть числа запятой, а не точкой. Кроме того, он может случайно ввести число словами (например, не «5», а «пять») — в этом случае слово нужно игнорировать.

**Ваша задача** — вывести минимум и максимум последовательности. Гарантируется, что пользователем будет введено хотя бы одно число.

Для нахождения минимума и максимума напишите функцию `find_min_max`, которая будет возвращать два числа.

**Пример:**

```
Ввод:
-2,56 25,002 восемь 19,13 13 -37,5

Вывод:
Minimum: -37.5
Maximim: 25.002
```



**Подсказка №1:**

Для проверки того, состоит ли строка `s` только из букв, можно воспользоваться методом `s.isalpha()`.

**Подсказка №2:**

Для поиска минимума и максимума в списке можно воспользоваться встроенными функциями `min()` и `max()`.

**Критерии оценивания:**

- Задача решена корректно, учтены все условия (**4 балла**):
    - программа корректно обрабатывает ввод пользователя и работает на любых входных данных;
    - программа корректно обрабатывает присутствие слов во вводе пользователя;
    - программа корректно рассчитывает минимум и максимум числовой последовательности;
    - решение сопровождено комментариями.
- Задача решена корректно, но не выполнено одно из условий (**2 балл**):
    - программа некорректно обрабатывает ввод пользователя и работает только на конкретном примере входных данных;
    - программа некорректно обрабатывает присутствие слов во вводе пользователя, что приводит к ошибкам при выполнении программы;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): программа некорректно рассчитывает минимум и максимум числовой последовательности.

````python
input_string = input('Введите последовательность чисел: ')


# создаем функцию для поиска мин и макс значения в строке
def find_min_max(str_1):
  
  
  # try-except, если при вводе будут поданы данные без чисел (например "восемь восемь восемь")
  try:

    # выбираем все элементы, которые не являются буквами и записываем их в переменную
    str_1 = ''.join(filter(lambda x: not x.isalpha(),  str_1))

    # заменяем запятые на точки и делим строку по пробелам (получаем список из строк)
    str_1 = str_1.replace(',', '.').split(' ')

    # удаляем пустые строки из списка, если они там есть
    while '' in str_1:str_1.remove('')

    # преобразуем все строки в списке str_1 в значения типа float
    str_1 = list(map(float, str_1))

    # выводим максимальное и минимальное значение списка
    return (f'Minimum: {min(str_1)} \nMaximim: {max(str_1)}')

  # вместо ошибки выводим надпись "Введены некорректные данные" и None
  except ValueError: return print('Введены некорректные данные')


print(find_min_max(input_string))
````

## Задание №3

#### Найти медиану

**Компания:** «Авито»

**Уровень:** Лёгкий

---
Медиана, как и среднее, является одной из важнейших характеристик для числовых данных в статистике.

**Медиана** — это число, которое находится в середине упорядоченной по возрастанию числовой последовательности.

Например, в последовательности чисел `1, 5, 2, 3, 6` медианой является число `3`, так как если упорядочить числа по возрастанию (`1, 2, 3, 5, 6`), центральным элементом в последовательности будет именно `3`.

**Примечание.** *Если в выборке чётное число элементов, то за медиану принимается среднее между центральными элементами (после упорядочивания).*

Например, в последовательности чисел `100, 5, 2, 4, 3, 6` медианой является число `4.5`, так как, если упорядочить числа по возрастанию (`2, 3, 4, 5, 6, 100`), то центральным элементом в последовательности будет среднее арифметическое между 4 и 5 ($\frac{4 + 5} {2} = 4.5$).

---
Пользователь с клавиатуры (через функцию `input()`) вводит через запятую и пробел неупорядоченную последовательность чисел.

Необходимо конвертировать ввод пользователя в список действительных чисел и вывести на экран медиану этой числовой последовательности.

Для поиска медианы в списке напишите функцию `find_median()`, которая будет возвращать одно число — медианное значение. Результат работы функции нужно вывести на экран.

Если переданная в функцию последовательность окажется пустой (пользователь ничего не ввёл) или пользователь ввёл числа некорректно (ввод нельзя преобразовать в числа), на экран должна быть выведена фраза `Некорректный ввод`.

**Пример №1:**
```
Ввод:
1, 5, 2, 3, 6

Вывод:
Median: 3.0
```
```
Ввод:
100, 5, 2, 4, 3, 6

Вывод:
Median: 4.5
```
```
Ввод:

Вывод:
Некорректный ввод
```
```
Ввод:
десять, 10, пять, 7, семь

Вывод:
Некорректный ввод
```


**Критерии оценивания:**

- Задача решена корректно, выполнены все условия (**6 балла**):
    - ввод пользователя конвертируется в список из чисел;
    - функция для расчёта медианы работает корректно для списка чётной и нечётной длины;
    - в случае некорректного ввода функция возвращает `None`;
    - решение сопровождено комментариями.
- Задача решена корректно, однако не выполнено одно из условий (**3 балл**):
    - ввод пользователя не конвертируется в список;
    - функция для расчёта медианы некорректно работает для списка чётной или нечётной длины;
    - в случае некорректного ввода функция отрабатывает с ошибкой;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): представлен некорректный алгоритм расчёта медианы для обоих случаев, результат работы программы зависит от вводимой последовательности чисел и т. д.

````python
input_string = input('Введите последовательность чисел: ')

# создаем рекурсивную функцию, которая будет определять медиану

def find_median(numbers):
  # создаем try-except для некорректных данных
  try:
    
    # разделяем строку на подстроки и записываем в список number; конвертируем их в int
    if type(numbers) is str: 
      numbers = numbers.split(', ')  
      numbers = list(map(int, numbers))     
      # сортируем по возрастнию
      numbers.sort() 

    else: # если number является списком
      
      # если длинна списка равна 1, возвращаем первый (и единственный) элемент списка
      if len(numbers) == 1: 
        return f'Median: {float(numbers[0])}'
      
      # если длинна списка равна 2, возвращаем среднее значение оставшихся элементов
      elif len(numbers) == 2: 
        return f'Median: {float((numbers[0]+numbers[1])/2)}'

    
    # если длинна списка number больше 2, "отрезаем" первый и последний элемент списка
    # и запускаем новый виток функции
    return find_median(numbers[1: len(numbers)-1])

  
  # если возникла либо ошибка IndexError или ValueError
  # функция выводит надпись 'Некорректный ввод', а также возвращает None
  except (IndexError, ValueError):print('Некорректный ввод')

print(find_median(input_string))
````

````python
number_word_dict = {
    "ты": 1000, "м": 1000000,
    "сто": 100, "двес": 200, "трис": 300, "четырес": 400, "пятьс": 500, "шестьс": 600, "семьс": 700, "восемьс": 800, "девятьс": 900,
    "одинн": 11, "двен": 12, "трин": 13, "четырн": 14, "пятн": 15, "шестн": 16, "семн": 17, "восемн": 18, "девятн": 19,
    "двад": 20, "трид": 30, "сор": 40, "пятьд": 50, "шестьд": 60, "семьд": 70, "восемьд": 80, "девяно": 90,
    "дес": 10, "н": 0, "о": 1, "дв": 2, "т": 3, "ч": 4, "п": 5, "ш": 6, "с": 7, "в": 8, "д": 9, }

input_string = input('Введите число словами: ')


# создаем функцию, состоящую из трех циклов
def transform_string_to_integer(s):
  s = s.lower() # приводим к нижнему регистру
  s = s.split(' ') # разделяем нашу строку по пробелам
  words_to_numbers = [] # создаем пустой список, в котором будут храниться слова, преобразованные в числа

  # проходим по словам в списке s и по содержимому словаря number_word_dict
  for word in s: 
    for i, j in number_word_dict.items():  
      
      # если ключ словаря сопадает с переменной word, записываем его в список words_to_numbers
      if word.startswith(i):     
        words_to_numbers.append(j)
        break 
      
  # вводим список из двух значений, которые будут выступать множителями    
  multipliers = [1000, 1000000] 
  
  # пустой список, в котором будут храниться все значения
  numbers = [] 
  
  # пустой список, для работы двух последующих циклов
  temp_lst = []  

  # проходимся по элементам ранее созданного списка words_to_numbers
  # если значение не является множителем, записываем в temp_lst
  for item in words_to_numbers:
    if item not in multipliers: 
      temp_lst.append(item) 
      
      
# если мы наткнулись на множитель, складываем содержимое temp_lst; добавляем их их сумму в numbers
# добавляем в тот же список множитель и очищаем temp_lst
    else:                       
      numbers.append(sum(temp_lst)) 
      numbers.append(item)   
      temp_lst = [] 
        
  numbers.append(sum(temp_lst)) # добавляем в numbers то, что в tem_lst было записано после последнего множителя




  # следующий цикл работает по аналогичному принципу
  # очищаем temp_lst
  temp_lst = [] 
  
  # создаем переменную, в которую будем записывать финальное значение
  final_count = 0 

  # проходимся по элементам списка numbers
  for i in numbers:  
    
    # если это значение не является множителем, записываем его в temp_lst
    if i not in multipliers:  
      temp_lst.append(i)  

    # если мы наткнулись на множитель, 
    # добавляем множитель в temp_lst; переменожаем содержимое temp_lst; прибавляем значение к final_count
    # очищаем temp_lst
    else:           
      temp_lst.append(i) 
      final_count += temp_lst[0]*temp_lst[1] 
      temp_lst = [] 

  final_count += temp_lst[0] # прибавляем к переменной final_count то, что в tem_lst было записано после последнего множителя

  return final_count # возвращаем final_count



print(transform_string_to_integer(input_string))


# получилось сделать так, чтобы функция работала в диапазоне от 0 до 999999999
````

### Задание №5 (бонусная, не оценивается)

#### Разложить факториал на простые множители

**Компания:** «СБЕР»

**Уровень:** Сложный

___

Пользователь с клавиатуры (через функцию `input()`) вводит некоторое целое число `n`.

---

Напомним, что **факториал числа $n$ ($n!$)** — это произведение всех натуральных чисел от $1$ до $n$ включительно.

Например, факториал числа $3$ будет равен: $3! = 1 * 2 * 3 = 6$.

---

**Простое число** — число, которое делится (нацело) только на себя и на 1. Например, $13$ — простое число, так как оно делится только на себя (на $13$) и на $1$, а $14$ не является простым числом, так как оно делится на $7$ и на $2$.

Подробнее о простых числах можно почитать [здесь](https://dzen.ru/a/XqFMe_VWclT8NTss).

---
**Простые множители** — это простые числа (не считая 1), на которые число делится без остатка.

Например, для числа $10$ простыми множителями являются числа $5$ и $2$. Для числа $28$ множителями являются числа $7$ и $4$, но простыми множителями будут числа $7$ и $2$, так как $4$ — это $2 * 2$, а число $28$ представляется в виде $28 = 7 * 2^2$.

----

В результате своей работы функция должна возвращать строку вида (приведён пример для `n=12`):

`12! = 2^10 * 3^5 * 5^2 * 7 * 11`

Здесь число слева от ^ — множитель, число справа от ^ — его степень. При этом если множитель встречается один раз, знак степени не выводится. Вывод множителей должен производиться в порядке их возрастания.

**Подсказка №2:**

Алгоритм нахождения простых множителей числа вы можете найти [здесь](https://foxford.ru/wiki/informatika/razlozhenie-chisla-na-mnozhiteli-v-python).

````python
# приводим ввод к типу int
n = int(input())

# создаем функциию
def decompose_factorial(n):

  # создаем внутреннюю функцию для определения значения факториала
  def factorial(n):
    if n == 1 or n == 0: return 1
    else: return n * factorial(n-1)

  # записываем в переменную fact_n значение факториала
  fact_n = factorial(n)

  # присваиваем делителю значение 2 (минимальный простой делитель)
  devider = 2
  
  # создаем пустой список, в который будем записывать последовательность простых множетелей
  lst = []

  # создаем цикл while, который будет принимать значение переменной fact_n и работать дор тех пор,
  # пока переменная fact_n > 1
  while fact_n > 1:
    
    # если fact_n делится на devider без остатка
    # присваиваем переменной fact_n значения остатка от деления на devider и записываем его в lst
    if fact_n % devider == 0:
      fact_n = fact_n // devider
      lst.append(devider)
      
    # если fact_n не делится на devider, увеличиваем его на 1
    else: devider += 1


  # создаем список count_lst и записываем в него кортежи из двух значений, в которых
  # первое значение - это делитель, а второе значение - это его частота в lst
  count_lst = list(map(lambda x: (x, lst.count(x)), set(lst)))


  # создаем пустую строку  final_lst
  final_str = ''
  
  # проходимся по значениям (кортежам) count_lst
  for i in count_lst:
     
    # если второе значение в кортеже больше 1 записываем первое значение, знак степени, и знак умножения
    if i[1] > 1:
      final_str += f'{i[0]}^{i[1]} * '
      
    # если второе значение в кортеже равно 1 записываем первое значение, и знак умножения
    else:
      final_str += f'{i[0]} * '

  # выводим получившуюся строку final_str без последних трех элементов (" * ")
  return f'{n}! = {final_str[:-3]}'


print(decompose_factorial(n))
````

**Подсказка:**

При решении задачи вам может понадобиться латинский алфавит. Вы можете получить его из встроенной в Python библиотеки `string`:

````python
import string
alphabet = string.ascii_lowercase
````

````python
s = input('Введите строку: ')
import string
alphabet = string.ascii_lowercase

# создаем рекурсивную функцию, которая будет удалять дубликаты
def replace_duplicates(s, step_count = 0): # на фход ей будут подаваться две переменных:
                                           # строка и переменная, которая будет использоваться для подсчета циклов
                                           # по дефолту 0

  # функция counter потребуется для подсчета циклов рекурсии
  def counter(count = 0):
    count += 1
    return count

  # для начала записываем строку в переменную string_lenth,
  # которая потребуется в нижеследующем цикле while
  string_lenth = s

  # задаем условие выхода из рекурсии (в строке s остались только уникальные элементы)
  if len(s) == len(set(s)): return f'output: {s}'

  # если это уссловие не выполняется
  else:
    # создаем переменную count и присваиваем значение 0
    count = 0
    # создаем цикл while, который будет повторять итерации пока подаваемая строка не изменится
    while s == string_lenth:

      # если элемент (на первой итерации цикла под индексом 0) встречается в строке после этого элемента
      if s[count] in s[count + 1:]:

        # тогда мы прибавляем к нему элемент
        if s[count] == 'z':s += 'a' # элемент 'a', если поданный элемент является 'z'
        else: s += alphabet[alphabet.find(s[count]) + 1]  # или элемент, находящийся на одну позицию дальше в алфавите

        # так как нам нужно удалить два одинаковывых элемента, при этом первый - крайний слева, а второй - крайний справа

        # создаем две перменные x и y 
        x = count # в переменную x записываем индекс буквы, который равен переемнной count (крайний слева)
        y = s.rfind(s[count]) # в переменную y записываем индекс той же самой буквы, но если считать справа налево (крайний справа)

        # преобразуем строку s
        # оставляем в ней все элементы, но пропускаем x и y
        s = s[ :x] + s[x+1 : y] + s[y+1: ]

        step_count = counter(step_count) # увеличиваем счетчик циклов рекурсии на 1
        print(f'step {step_count}: ' + s) # выводим текущий номер цикла рекурсии и измененную строку

        # отправляем изменную строку s и измененный счетчик циклов на следующую итерацию рекурсии
        return replace_duplicates(s, step_count)

      # если элемент под индексом count не обнаружен в остатке строки
      # увеличиваем count на 1 и запускаем цикл while снова
      else: count += 1


print(replace_duplicates(s))
````
